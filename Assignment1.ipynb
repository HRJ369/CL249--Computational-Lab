```python
import numpy as np  # Import the NumPy library for numerical operations
#MY roll is 22b0450, so adding 1 in the last digit it is still 1.
# Initialize matrix A and vector b with given values
A = np.array([
    [52, 93, 15, 72, 61, 21, 83, 87, 75, 75, 88, 24],
    [3, 22, 53, 2, 88, 30, 38, 2, 64, 60, 21, 33],
    [76, 58, 22, 89, 49, 91, 59, 42, 92, 60, 80, 15],
    [62, 62, 47, 0, 51, 55, 64, 3, 51, 7, 21, 73],
    [39, 18, 4, 89, 60, 14, 9, 90, 53, 2, 84, 92],
    [60, 71, 44, 8, 47, 35, 78, 81, 36, 50, 4, 2],
    [6, 54, 4, 54, 93, 63, 18, 90, 44, 34, 74, 62],
    [14, 95, 48, 15, 72, 78, 87, 1, 40, 85, 80, 82],
    [53, 24, 26, 89, 60, 41, 29, 15, 45, 65, 89, 71],
    [9, 88, 1, 8, 88, 63, 11, 81, 8, 35, 35, 33],
    [5, 41, 28, 7, 73, 72, 12, 34, 33, 48, 23, 62],
    [88, 37, 99, 44, 86, 91, 35, 65, 99, 47, 78, 3]
], dtype=float)  # Define a 12x12 matrix A with float data type

b = np.ones(12, dtype=float)  # Initialize vector b with ones

def gaussian_elimination(A, b):
    n = len(b)  # Get the length of vector b
    ops_count = 0  # Initialize operation counter

    for k in range(n - 1):  # Loop through each column
        # Find the row with the maximum element in the current column
        max_row = max(range(k, n), key=lambda i: abs(A[i, k]))
        if max_row != k:  # Swap the current row with the row having the maximum element
            A[[k, max_row]] = A[[max_row, k]]  # Swap rows in matrix A
            b[[k, max_row]] = b[[max_row, k]]  # Swap corresponding elements in vector b
            ops_count += 2 * n  # Increment operation count

        for i in range(k + 1, n):  # Eliminate entries below the pivot
            if A[i, k] != 0:  # If the element is not zero
                factor = A[i, k] / A[k, k]  # Calculate the factor to eliminate the entry
                ops_count += 1  # Increment operation count
                for j in range(k, n):  # Update the row elements
                    A[i, j] -= factor * A[k, j]  # Subtract the scaled row
                    ops_count += 2  # Increment operation count
                b[i] -= factor * b[k]  # Update the corresponding element in vector b
                ops_count += 2  # Increment operation count

    x = np.zeros(n)  # Initialize the solution vector with zeros
    for i in range(n - 1, -1, -1):  # Perform back substitution
        x[i] = (b[i] - np.dot(A[i, i + 1:], x[i + 1:])) / A[i, i]  # Calculate the solution for the current variable
        ops_count += 2 + (n - i - 1) * 2  # Increment operation count

    return x, A, ops_count  # Return the solution vector, the upper triangular matrix, and the operation count

# Solve the system using Gaussian elimination
solution, upper_triangular, ops_count = gaussian_elimination(A.copy(), b.copy())

# Calculate the determinant of the upper triangular matrix
det = np.prod(np.diag(upper_triangular))

# Print the upper triangular matrix
print("Upper Triangular Matrix:")
print(upper_triangular)

# Print the determinant of the matrix
print("\nDeterminant:", det)

# Print the solution vector
print("\nSolution Vector:")
print(solution)

# Print the total number of operations performed
print("\nTotal Operations:", ops_count)

# Verify the solution by calculating A * x
A_times_x = np.dot(A, solution)
print("\nA * x:")
print(A_times_x)

# Print the original b vector
print("\nOriginal b vector:")
print(np.ones(12))
```

```python
import numpy as np

# Define the composition of the feed (as fractions) and the total feed flow rate
feed_comp = np.array([0.15, 0.25, 0.40, 0.20])
total_feed = 70  # kmol/hr

# Define the compositions (as fractions) in the streams
comp_D1 = np.array([0.07, 0.04, 0.54, 0.35])
comp_B1 = np.array([0.18, 0.24, 0.42, 0.16])
comp_D2 = np.array([0.15, 0.10, 0.54, 0.21])
comp_B2 = np.array([0.24, 0.65, 0.10, 0.01])

# Build the coefficient matrix based on component balances
coeff_matrix = np.array([
    [comp_D1[0], comp_B1[0], comp_D2[0], comp_B2[0], 0, 0],  # Xylene balance
    [comp_D1[1], comp_B1[1], comp_D2[1], comp_B2[1], 0, 0],  # Styrene balance
    [comp_D1[2], comp_B1[2], comp_D2[2], comp_B2[2], 0, 0],  # Toluene balance
    [comp_D1[3], comp_B1[3], comp_D2[3], comp_B2[3], 0, 0],  # Benzene balance
    [1, 1, 0, 0, -1, 0],  # D1 + B1 - D = 0
    [0, 0, 1, 1, 0, -1],  # D2 + B2 - B = 0
], dtype=float)

# Create the right-hand side vector based on the total feed and balance equations
rhs = np.array([
    total_feed * feed_comp[0],  # Total Xylene in feed
    total_feed * feed_comp[1],  # Total Styrene in feed
    total_feed * feed_comp[2],  # Total Toluene in feed
    total_feed * feed_comp[3],  # Total Benzene in feed
    0,  # D1 + B1 - D = 0
    0   # D2 + B2 - B = 0
], dtype=float)

# Combine the coefficient matrix with the right-hand side vector to create the augmented matrix
aug_matrix = np.hstack([coeff_matrix, rhs.reshape(-1, 1)])

# Function to perform Gauss-Jordan Elimination
def gauss_jordan(matrix):
    num_rows, num_cols = matrix.shape

    for i in range(num_rows):
        # Pivot element in the current row
        pivot = matrix[i, i]

        # If pivot is zero, swap with a row below that has a non-zero element in the same column
        if pivot == 0:
            for k in range(i+1, num_rows):
                if matrix[k, i] != 0:
                    matrix[[i, k]] = matrix[[k, i]]
                    pivot = matrix[i, i]
                    break

        # Normalize the pivot row
        matrix[i] = matrix[i] / pivot

        # Eliminate the current column elements in other rows
        for j in range(num_rows):
            if i != j:
                factor = matrix[j, i]
                matrix[j] -= factor * matrix[i]

    # The solution is the last column of the augmented matrix
    return matrix[:, -1]

# Apply Gauss-Jordan elimination to the augmented matrix
sol = gauss_jordan(aug_matrix.copy())

# Extract the results and print
D1, B1, D2, B2, D, B = sol
print(f"Distillate 1: {D1:.4f} kmol/hr")
print(f"Bottoms 1: {B1:.4f} kmol/hr")
print(f"Distillate 2: {D2:.4f} kmol/hr")
print(f"Bottoms 2: {B2:.4f} kmol/hr")
print(f"Distillate: {D:.4f} kmol/hr")
print(f"Bottoms: {B:.4f} kmol/hr")
```
