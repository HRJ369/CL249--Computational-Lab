```python
#SLIDE 1
b = float(input("Enter the number: "))  # Number to find the square root of
x_old = b  # Initial guess for the square root

while True:
    x_new = 0.5 * (x_old + b / x_old)  # Babylonian method formula
    if abs(x_new - x_old) < 0.05:  # Convergence criterion
        break
    x_old = x_new  # Update x_old for the next iteration

print(f"Approximation of the square root is: {x_new}")
```
