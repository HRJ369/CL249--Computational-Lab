```python
import matplotlib.pyplot as plt

# Given data points
x = [2495.09, 3742.63, 4990.17]
y = [15.2386, 15.0815,15.032]

# Plotting the points and connecting them with a line
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Labeling the axes
plt.xlabel('Reynolds Number')
plt.ylabel('Nusselt Number')

# Title of the plot
plt.title('Nu vs Re')

# Displaying the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data points
x = np.array([3.397085722, 3.573176981, 3.698115718])
y = np.array([1.182943711,1.17884, 1.17702])

# Perform linear regression to fit a line to the data
slope, intercept = np.polyfit(x, y, 1)

# Create a range of x values for plotting the fitted line
x_fit = np.linspace(min(x), max(x), 100)
y_fit = slope * x_fit + intercept

# Plot the scatter plot
plt.scatter(x, y, color='r', label='Data points')

# Plot the linear fit
plt.plot(x_fit, y_fit, color='b', label=f'Fit: y = {slope:.4f}x + {intercept:.4f}')

# Labeling the axes
plt.xlabel('Log Re')
plt.ylabel('Log Nu')

# Title of the plot
plt.title('Log Log plot of Nusselt and Reynolds Number with Linear Fit')

# Display the legend
plt.legend()

# Display the plot
plt.show()
```
