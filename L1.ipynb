```python
#BABYLONIAN WAY TO FIND THE SQUARE ROOT OF A NUMBER A
import math as m    #Library which is used to to find square root normally
a = int(input("Enter the number = "))  #taking input of the number whose root we are interested in
x = int(input("Enter the initial guess = "))            #Our initial guess
n=int(input("Enter the number of iterations = "))              #No of iterations we want it to go
for i in range (0,n):                                 #loop
  x=(x/2)+ (a/(2*x))       #updation

print("Here is the estimated root by babylonian method", x)     #the root using babylonian method
print("Here is the actual root ", m.sqrt(a))                       #actual root

```

```python
#CRE, FINDING THE ACTIVATION ENERGY
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
T =[313,319,323,328,333]
k= [0.00043, 0.00103,0.00180,0.00355,0.00717]
Inverse_T = [1/t for t in T if t!=0 ]
lnk = [m.log(i) for i in k if k!=0]
plt.plot(Inverse_T,lnk)
plt.xlabel("1/T")
plt.ylabel("ln(k)")
plt.title('CRE PLOT')
plt.show()
slope = (lnk[2]-lnk[3])/(Inverse_T[2]-Inverse_T[3])
print("The activation energy is ", slope/8.314)
```
