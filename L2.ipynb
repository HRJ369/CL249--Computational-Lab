```python
import numpy as np

# Initialize matrix A and vector b
A = np.array([
    [52, 93, 15, 72, 61, 21, 83, 87, 75, 75, 88, 24],
    [3, 22, 53, 2, 88, 30, 38, 2, 64, 60, 21, 33],
    [76, 58, 22, 89, 49, 91, 59, 42, 92, 60, 80, 15],
    [62, 62, 47, 0, 51, 55, 64, 3, 51, 7, 21, 73],
    [39, 18, 4, 89, 60, 14, 9, 90, 53, 2, 84, 92],
    [60, 71, 44, 8, 47, 35, 78, 81, 36, 50, 4, 2],
    [6, 54, 4, 54, 93, 63, 18, 90, 44, 34, 74, 62],
    [14, 95, 48, 15, 72, 78, 87, 1, 40, 85, 80, 82],
    [53, 24, 26, 89, 60, 41, 29, 15, 45, 65, 89, 71],
    [9, 88, 1, 8, 88, 63, 11, 81, 8, 35, 35, 33],
    [5, 41, 28, 7, 73, 72, 12, 34, 33, 48, 23, 62],
    [88, 37, 99, 44, 86, 91, 35, 65, 99, 47, 78, 3]
], dtype=float)

b = np.ones(12, dtype=float)

def gaussian_elimination(A, b):
    n = len(b)
    ops_count = 0

    for k in range(n - 1):
        max_row = max(range(k, n), key=lambda i: abs(A[i, k]))
        if max_row != k:
            A[[k, max_row]] = A[[max_row, k]]
            b[[k, max_row]] = b[[max_row, k]]
            ops_count += 2 * n

        for i in range(k + 1, n):
            if A[i, k] != 0:
                factor = A[i, k] / A[k, k]
                ops_count += 1
                for j in range(k, n):
                    A[i, j] -= factor * A[k, j]
                    ops_count += 2
                b[i] -= factor * b[k]
                ops_count += 2

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (b[i] - np.dot(A[i, i + 1:], x[i + 1:])) / A[i, i]
        ops_count += 2 + (n - i - 1) * 2

    return x, A, ops_count

solution, upper_triangular, ops_count = gaussian_elimination(A.copy(), b.copy())

det = np.prod(np.diag(upper_triangular))

print("Upper Triangular Matrix:")
print(upper_triangular)
print("\nDeterminant:", det)
print("\nSolution Vector:")
print(solution)
print("\nTotal Operations:", ops_count)

A_times_x = np.dot(A, solution)
print("\nA * x:")
print(A_times_x)
print("\nOriginal b vector:")
print(np.ones(12))

```

```python
import numpy as np

# Feed composition (as fractions)
F_xylene = 0.15
F_styrene = 0.25
F_toluene = 0.40
F_benzene = 0.20
F_total = 70  # kmol/hr

# Compositions (as fractions) in the streams
D1_composition = np.array([0.07, 0.04, 0.54, 0.35])
B1_composition = np.array([0.18, 0.24, 0.42, 0.16])
D2_composition = np.array([0.15, 0.10, 0.54, 0.21])
B2_composition = np.array([0.24, 0.65, 0.10, 0.01])

# Total balance equations
# D + B = F
# D1 + B1 = D
# D2 + B2 = B

# Component balance equations for Xylene, Styrene, Toluene, and Benzene

A = np.array([
    [D1_composition[0], B1_composition[0], D2_composition[0], B2_composition[0], 0, 0],
    [D1_composition[1], B1_composition[1], D2_composition[1], B2_composition[1], 0, 0],
    [D1_composition[2], B1_composition[2], D2_composition[2], B2_composition[2], 0, 0],
    [D1_composition[3], B1_composition[3], D2_composition[3], B2_composition[3], 0, 0],
    [1, 1, 0, 0, -1, 0],  # D1 + B1 - D = 0
    [0, 0, 1, 1, 0, -1],  # D2 + B2 - B = 0
])

b = np.array([
    F_total * F_xylene,
    F_total * F_styrene,
    F_total * F_toluene,
    F_total * F_benzene,
    0,  # D1 + B1 - D = 0
    0   # D2 + B2 - B = 0
])

# Solving the system of equations
solution = np.linalg.solve(A, b)

# Output results
D1, B1, D2, B2, D, B = solution
print(f"D1: {D1:.4f} kmol/hr")
print(f"B1: {B1:.4f} kmol/hr")
print(f"D2: {D2:.4f} kmol/hr")
print(f"B2: {B2:.4f} kmol/hr")
print(f"D : {D:.4f} kmol/hr")
print(f"B : {B:.4f} kmol/hr")

```
